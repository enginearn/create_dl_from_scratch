## Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
print(sigmoid(x))
```

## Relu: Rectified Linear Unit

$$
h(x) = \begin{cases}
       x & \text{if } x \gt 0 \\
       0 & \text{if } x \leq 0
       \end{cases}
$$

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.0) # y軸の範囲を指定
plt.show()
```

```python
y1 = sigmoid(x)
y2 = step_function_np(x)
y3 = relu(x)

plt.plot(x, y1, label='sigmoid(x)')
plt.plot(x, y2, linestyle='--', label='step_function_np(x)')
plt.plot(x, y3, linestyle=':', label='relu(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

1層目の出力を$A^{(1)}$とすると、$A^{(1)}$は以下のように表せる。

$$
    a_1^{(1)} = w_{11}^{(1)}x_1 + w_{12}^{(1)}x_2 + b_1^{(1)}
$$

$$
    A^{(1)} = W^{(1)}X + b^{(1)}
$$

$$
    A^{(1)} = 
    \begin{bmatrix}
        w_{11}^{(1)} & w_{12}^{(1)} \\
        w_{21}^{(1)} & w_{22}^{(1)} \\
        w_{31}^{(1)} & w_{32}^{(1)} \\
    \end{bmatrix}
    \begin{bmatrix}
        x_1 \\
        x_2 \\
    \end{bmatrix}
    +
    \begin{bmatrix}
        b_1^{(1)} \\
        b_2^{(1)} \\
        b_3^{(1)} \\
    \end{bmatrix}
$$

$$
    A^{(1)} = 
    \begin{bmatrix}
        w_{11}^{(1)}x_1 + w_{12}^{(1)}x_2 + b_1^{(1)} \\
        w_{21}^{(1)}x_1 + w_{22}^{(1)}x_2 + b_2^{(1)} \\
        w_{31}^{(1)}x_1 + w_{32}^{(1)}x_2 + b_3^{(1)} \\
    \end{bmatrix}
$$

```python
X = np.array([1.0, 0.5]) # input
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]]) # weight
B1 = np.array([0.1, 0.2, 0.3]) # bias

print(W1.shape) # (2, 3)
print(X.shape) # (2,)
print(B1.shape) # (3,)

A1 = np.dot(X, W1) + B1
print(A1) # [0.3, 0.7, 1.1]
```

## Input layer

```python
def init_network():
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])

    return network

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    # 1st layer
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)

    # 2nd layer
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)

    # 3rd layer
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)

    return y
```

```python
network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```

## Output layer

> ニューラルネットワークは、分類問題と回帰問題の両方に用いることができます。
> ただし、分類問題と回帰問題では、出力層の活性化関数が異なります。
> 一般的には、分類問題ではソフトマックス関数、回帰問題では恒等関数を使います。
>
> 機械学習の問題は、分類問題と回帰問題のどちらかになります。
> 分類問題とは、ある入力データをあらかじめ決められたカテゴリに分類する問題です。
> 例えば、画像に写っている人の写真を「男性」「女性」のどちらかに分類する問題です。
> 一方、回帰問題とは、ある入力データから数値を予測する問題です。
> 例えば、ある人の年齢を予測する問題です。

### Softmax Function

$$
    y_k = \frac{e^{a_k}}{\sum_{i=1}^{n} e^{a_i}}
$$

$$
\begin{align}
\hat{y} &= softmax(XW + b) \\
\end{align}
$$

```python
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a) # exponential function
print(exp_a)

sum_exp_a = np.sum(exp_a) # sum of exponential function
print(sum_exp_a)

y = exp_a / sum_exp_a # softmax function
print(y)
```

```python
def softmax(x):
    c = np.max(x)
    exp_x = np.exp(x - c)
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    return y
```

```python
x = np.array([1010, 1000, 990])
y = softmax(x)
print(y)
```

## Inference processing

```python
import pickle
```

```python
with open("../dataset/sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
        print(network.keys())
```

```python
def get_data():
    (x_train, t_train) , (x_test, t_test) = \
        load_mnist(normalize=True, flatten=True, one_hot_label=False)
    return x_test, t_test
```

```python
def init_network():
    with open("../dataset/sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    return network
```

```python
def predict(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]

    # Forward propagation
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y
```

```python
x, t = get_data()
network = init_network()

accuracy_cnt = 0
for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y) # get the index of the highest value
    if p == t[i]:
        accuracy_cnt += 1

print("Accuracy: " + str(float(accuracy_cnt) / len(x)))
```

```python
batch_size = 100
accuracy_cnt = 0

for i in range(0, len(x), batch_size):
    x_batch = x[i:i + batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum(p == t[i:i + batch_size])

print("Accuracy: " + str(float(accuracy_cnt) / len(x)))
```
